### Calculate angles and radii of leaves and petals

```python
# import golden angle
# set to radians

golden_angle = 137.5077640500378546463487396283702776206886952699253696312384958261062333851951
phi = golden_angle * ( math.pi / 180.0 ) 


# calculate angles and radii of phyllotactic points beforehand
thetas = []
radii = []

for n in range(200):

    r = math.sqrt(n) 
    theta = n * phi 
    
    thetas.append(theta)
    radii.append(r)
```

### Create animation

```python
# This loop is selecting a growing number of leaves
for i in range(200): # for 200 leaves

    if i%20==0: # print if divisible by 20
        print(i)

    # For animation, you need to call a figure
    fig = plt.figure(figsize=(10,10), facecolor="white")  

    # creates x, y coordinates for each leaf/petal
    xlist = []
    ylist = []
    
    # This selects the current set of thetas to use
    current_thetas = thetas[0:(i+1)]
    # This selects the current set of radii to use
    # Everytime the loop runs, this list will be recreated in the "forward" direction
    current_radii = radii[0:(i+1)]
    # After creating the list of radii, "revere" its order
    # Reversing the order of the radii means that the 
    # leaves first specified have the longest radius
    current_radii.reverse()

    # In this loop, for each of the current leaves, calculate x & y coordinates
    for n in range (i+1): 
        
        # Here, for each leaf calculate n caclulate r, theta, x, y, and append x & y to their lists
        # theta & r are calcualted by indexing using n the lists current_thetas and current_radii
        # x & y are calculated from theta & r, and then appended to xlist and ylist
        r = current_radii[n]
        theta = current_thetas[n]
                  
        x = r * math.cos(theta)
        y = r * math.sin(theta)
            
        xlist.append(x)
        ylist.append(y)

    # set the alpha for each leaf
    for i in range(len(current_radii)):
        
        if current_radii[i] > 4: # create a maximum leaf length
            leaf_r = 4
        else:
            leaf_r = current_radii[i]
            
        if current_radii[i] > 5: # this is for old leaves to disappear using alpha
            alfa = 0
        elif current_radii[i] > 4.8:
            alfa = 0.1
        elif current_radii[i] > 4.7:
            alfa = 0.2
        elif current_radii[i] > 4.6:
            alfa = 0.3
        elif current_radii[i] > 4.5:
            alfa = 0.4
        elif current_radii[i] > 4.4:
            alfa = 0.5
        elif current_radii[i] > 4.3:
            alfa = 0.6
        elif current_radii[i] > 4.2:
            alfa = 0.7
        elif current_radii[i] > 4.1:
            alfa = 0.8
        elif current_radii[i] > 4.0:
            alfa = 0.9
        else:
            alfa=1

        leaf_theta = current_thetas[i]

        ###############################
        ### ROTATE, SCALE, AND PLOT ###
        ###############################

        ##############
        #### AGAVE ###
        ##############
        
        xaf = []
        yaf = []

        for j in range(len(xa)): # for each leaf, rotate and scale
            
            xff = leaf_r*(xa[j]*math.cos(leaf_theta) - ya[j]*math.sin(leaf_theta))
            yff = leaf_r*(ya[j]*math.cos(leaf_theta) + xa[j]*math.sin(leaf_theta))
            
            xaf.append(xff)
            yaf.append(yff)

        ###################
        #### AGAVE PLOT ###
        ###################
    
        plt.subplot(2,2,1)
        plt.fill(xaf, yaf, color="#6D9DB3", alpha=alfa/2, zorder=i)
        plt.plot(xaf, yaf, color="white", alpha=alfa/2, lw=1, zorder=i)
        plt.xlim(-4,4)
        plt.ylim(-4,4)
        
        plt.gca().set_aspect('equal')
        plt.axis('off')

        ###################
        ### CEMPASÚCHIL ###
        ###################

        # Petal
        xcpf = []
        ycpf = []

        for j in range(len(xcp)): # for each leaf, rotate and scale
            
            xff = leaf_r*(xcp[j]*math.cos(leaf_theta) - ycp[j]*math.sin(leaf_theta))
            yff = leaf_r*(ycp[j]*math.cos(leaf_theta) + xcp[j]*math.sin(leaf_theta))
            
            xcpf.append(xff)
            ycpf.append(yff)

        # Tip
        xctf = []
        yctf = []

        for j in range(len(xct)): # for each leaf, rotate and scale
            
            xff = leaf_r*(xct[j]*math.cos(leaf_theta) - yct[j]*math.sin(leaf_theta))
            yff = leaf_r*(yct[j]*math.cos(leaf_theta) + xct[j]*math.sin(leaf_theta))
            
            xctf.append(xff)
            yctf.append(yff)

        #########################
        #### CEMPASÚCHIL PLOT ###
        #########################

        plt.subplot(2,2,2)
        plt.fill(xcpf, ycpf, color="#FF5B00", alpha=alfa/1, zorder=i)
        plt.plot(xcpf, ycpf, color="#FFB700", alpha=alfa/1, zorder=i)
        plt.fill(xctf, yctf, color="#FF9C00", alpha=alfa/1, zorder=i)
        plt.xlim(-5,5)
        plt.ylim(-5,5)
        
        plt.gca().set_aspect('equal')
        plt.axis('off')

        #################
        ### SUCCULENT ###
        #################

        # Leaf
        xswf = []
        yswf = []

        for j in range(len(xsw)): # for each leaf, rotate and scale
            
            xff = leaf_r*(xsw[j]*math.cos(leaf_theta) - ysw[j]*math.sin(leaf_theta))
            yff = leaf_r*(ysw[j]*math.cos(leaf_theta) + xsw[j]*math.sin(leaf_theta))
            
            xswf.append(xff)
            yswf.append(yff)

        # Tip
        xstf = []
        ystf = []

        for j in range(len(xst)): # for each leaf, rotate and scale
            
            xff = leaf_r*(xst[j]*math.cos(leaf_theta) - yst[j]*math.sin(leaf_theta))
            yff = leaf_r*(yst[j]*math.cos(leaf_theta) + xst[j]*math.sin(leaf_theta))
            
            xstf.append(xff)
            ystf.append(yff)

        #######################
        #### SUCCULENT PLOT ###
        #######################

        plt.subplot(2,2,3)
        plt.fill(xswf, yswf, color="#38a28b", alpha=alfa/5)
        plt.fill(xstf, ystf, color="#fd3b5d", alpha=alfa/2)
        plt.xlim(-5,5)
        plt.ylim(-5,5)
        
        plt.gca().set_aspect('equal')
        plt.axis('off')

        #############
        ### YUCCA ###
        #############

        # Leaf
        x_yucf = []
        y_yucf = []

        for j in range(len(x_yuc)): # for each leaf, rotate and scale
            
            xff = leaf_r*(x_yuc[j]*math.cos(leaf_theta) - y_yuc[j]*math.sin(leaf_theta))
            yff = leaf_r*(y_yuc[j]*math.cos(leaf_theta) + x_yuc[j]*math.sin(leaf_theta))
            
            x_yucf.append(xff)
            y_yucf.append(yff)

        ##################
        ### YUCCA PLOT ###
        ##################

        plt.subplot(2,2,4)
        plt.fill(x_yucf, y_yucf, color="#94B447", alpha=alfa/3)
        plt.plot(x_yucf, y_yucf, color="#94B447", alpha=alfa, lw=1.3)
        plt.xlim(-4,4)
        plt.ylim(-4,4)
        
        plt.gca().set_aspect('equal')
        plt.axis('off')
 
    # Save the animation
    plt.tight_layout()
    plt.savefig("./dia_de_muertos/X_"+ str(i)+".png",facecolor=fig.get_facecolor(), edgecolor='none')

    # Closes the figure animation once complete
    plt.close()
   
```
